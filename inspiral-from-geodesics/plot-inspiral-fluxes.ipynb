```julia
using Plots
using DelimitedFiles
```

```julia
inspiralfluxes = DelimitedFiles.readdlm("inspiral-fluxes-v1.csv", ',', header=true)
fluxes = inspiralfluxes[1]

tfluxes = fluxes[:,1]
e = fluxes[:,2]
p = fluxes[:,3]
ι = fluxes[:,4]
Efluxes = fluxes[:,5]
Lfluxes = fluxes[:,6]
Qfluxes = fluxes[:,7]
```

```julia
plot(tfluxes, Efluxes,
    xlabel = "t",
    label = "Eflux")
plot!(tfluxes, Lfluxes,
    label = "Lflux")
plot!(tfluxes, Qfluxes,
    label = "Qflux")
```
